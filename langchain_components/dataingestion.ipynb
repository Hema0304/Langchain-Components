```python
#text loder
from langchain_community.document_loaders import TextLoader

loder = TextLoader('speech.txt')
documents = loder.load()
documents
```

```python
# reading a pdf file 
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader('sample_langchain.pdf')
docs=loader.load()
docs
```

```python
#web based loder

from langchain_community.document_loaders import WebBaseLoader
loader = WebBaseLoader(web_paths=['https://en.wikipedia.org/wiki/Artificial_intelligence',])
docs = loader.load()
docs
```

```python
from langchain_community.document_loaders import WebBaseLoader
import bs4
loader = WebBaseLoader(web_paths=('https://en.wikipedia.org/wiki/Artificial_intelligence',),
                                  bs_kwargs=dict(parse_only=bs4.SoupStrainer(
                                      class_=("mw-page-container","mw-page-container-inner")
                                  )))
docs = loader.load()
docs
```

```python
from langchain_community.document_loaders import ArxivLoader
loader = ArxivLoader(query="artificial intelligence", max_results=5)
docs = loader.load()
docs
```

```python
from langchain_community.document_loaders import WikipediaLoader
loader = WikipediaLoader(query="artificial intelligence",load_max_docs=5).load()
print(loader)
```
